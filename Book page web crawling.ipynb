```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

current_page = 1
max_pages = 3

while(current_page<=max_pages):

    page= requests.get('http://books.toscrape.com/catalogue/page-'+str(current_page)+'.html')
    soup= BeautifulSoup(page.content, 'html.parser')
    books = soup.find(class_='col-sm-8 col-md-9')

    items = books.find_all(class_ = 'product_pod')

    #print(items[0].div.a.img.get('alt'))
    #print(items[0].find(class_='price_color').get_text())


    book_name = [item.div.a.img.get('alt') for item in items]
    book_price = [item.find(class_= 'price_color').get_text() for item in items]
    
    df = pd.DataFrame({

    'Book_Name' : book_name,
    'Book_Price' : book_price

    })

    print(df)

    current_page = current_page+1 
    
    #print(book_price)
    #print(book_name)
    


    





```
